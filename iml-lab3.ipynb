```python
#name: rathod dhruvi rajubhai
#enrollment no:216270307090
#Practical No.3: Write a NumPy program to implement following operation 
#to stack arrays horizontally (column wise)
import numpy as np
a = np.array([[4,5,5],
            [3,3,3]])
b = np.array([[9, 8, 7],
                 [6, 5, 4]])
print("\na: \n", a)
print("\nb: \n", b)

# horizontally stack the arrays (axis = 1)
c = np.hstack([a, b])
print("\nc: \n", c)
 
```
